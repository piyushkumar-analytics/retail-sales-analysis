## Droping Columns

Few columns need to be droped that are not much necessary for the data

```python
df.drop(['Item', 'Discount Applied'], axis=1, inplace=True)
```


The **Item** column was removed because it mostly contained the same information as the **Category** column along with some extra numbers and text, making it redundant for the analysis.

The **Discount** column was also removed because it only contained True/False values and did not appear to be meaningfully applied in the dataset.

# Filling Missing Data

```python
df.isna().sum()
```

```python
df.describe()
```

```python
# Before we fill the missing data we need to make sure that the Total Spent column is correct.
 
df['Total Spent'] = df['Price Per Unit'] * df['Quantity']
```

```python
miss = ['Price Per Unit', 'Quantity', 'Total Spent']
df[miss] = df[miss].fillna(df[miss].median())
```

```python
df.isna().sum()
```

```python
# The columns Price Per Unit, Quantity, and Total Spent contain missing values.  
# To handle this issue, the median of each column is used to fill the missing values
# because it is less affected by outliers and better represents the typical value.
```

## Correcting Data Types

Some columns may have incorrect data types so we need to convert them into appropriate format.


```python
df.info()
```

```python
# For Category Type:
cat = ['Location','Category','Payment Method']
df[cat] = df[cat].astype('category')

# For int Type:
df['Quantity'] = df['Quantity'].astype('int')

# For Date Type:
df['Transaction Date'] =  pd.to_datetime(df['Transaction Date'], errors='coerce')
```
